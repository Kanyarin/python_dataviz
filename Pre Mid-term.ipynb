 ### สร้าง function ชื่อ createDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
- ถ้า True จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้า False จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น

เช่น

output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )

จะได้ output1 เป็น { 'apple' : 1 , 'banana' : 2 }

output2 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )

จะได้ output2 เป็น { 1 : 'apple' , 2 : 'banana' }
---จำการสร้างฟังก์ชัน/loop/if else ให้ได้ output แบบอาจารย์---

```python
def createDict(listtext, listnumber, logicV): #ถ้า text=2 / range = 0,1 / index=ตัวแปร
    if logicV : #logicv == True
        retdict = {} #สร้างdictเปล่าขึ้นมา
        for index in range(len(listtext)):  
            retdict[listtext[index]]= listnumber[index] #dictทีละตัว #text ก่อน number
    else:
        retdict = {} 
        for index in range(len(listtext)):  
             retdict[listnumber[index]]= listtext[index] #number djvo text
            
    return retdict  #ส่งค่าoutput
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True ) #True=logic
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False ) #T
```

#### How many time A, T, G, C in the sequence? ## A T G C มีอยู่กี่ตัว

```python
def countChar(seq,character): #countChar=ชื่อฟังก์ชันของตัวแปรที่จะเก็บ
    charcount = 0 #กำหนดตัวแปรไว้เก็บค่า
    for char in seq: #วนลูปจำนวนแถวในตาราง
        if char == character: #ถ้าตัวแปรที่ใส่มามีค่า=ตัวแปรที่เรากำหนด
            charcount = charcount + 1 #1ตัวที่ซ้ำ
    return charcount
```

```python
countChar('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGA','A')
```

```python
ANum = 0
TNum = 0
GNum = 0         #กำหนดตัวแปร
CNum = 0
for row in range(DNAseq.shape[0]):    #วนลูป
    ANum += countChar(DNAseq.iloc[row][0],'A')
    TNum += countChar(DNAseq.iloc[row][0],'T')       #มีกี่ตัวก็ให้ไปเก็บใน anum tnum gnum cnum / += คือบวกตัวมันเอง
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')
print(f'There are {ANum} As, {TNum} S, {GNum} Gs, {CNum} Cs')
```

### What is the most frequence 5-char-long subsequence

```python
'ATTAAAGGTTT'
```

```python
#ฟังก์ชันการตัดตัวหนังสือ
def populate5charstring(seq):
    str_list = list()   #สร้างlist
    for i in range(len(seq)-4): #ตัด 5(ที่ให้มา) เหลือ -4 / ตัด 20 เหลือ -19  #วนลูป
        str_list.append(seq[i:i+5]) #append เพิ้มสมาชิกตัวที่ i จนถึง i+5 -> i=1 : 1+5=6
    return str_list
```

```python
populate5charstring('ATTAAAGGTTT')   #(DNAseq.iloc[1][0])
```

```python
list_all_5char = list() #สร้าง list เพื่อเก็บ
for row in range(DNAseq.shape[0]):   #วนลูปตามจำนวนแถว
    list_all_5char = list_all_5char + populate5charstring(DNAseq.iloc[row][0])
```

```python
len(list_all_5char) #ใน list นี้ยังซ้ำกัน นำมาคำนวณไม่ได้ + ตัดไปแล้ว 5
```

```python
unique_5char = list(set(list_all_5char)) #set = คำสั่งตัดตัวซ้ำกันให้เอาออก
```

```python
len(unique_5char) #ตัวที่ไม่ซ้ำกันเลย เป็นยี่ห้อนมที่อยู่ในลัง ที่ไม่ซ้ำกันเลย 1022 ยี่ห้อจากนมทั้งหมด 28191 กล่องในลัง
```

```python
dict_5char = {}
for sub_str in unique_5char:
    num = 0
    for s in list_all_5char:
        if sub_str == s:
            num += 1   #ถ้า = ให้+1 ถ้าไม่เท่ากับ 5 จะเปลี่ยนเป็นตัวที่ 2 GATAA ของ 1022
    dict_5char[sub_str] = num
```

```python
dict_5char
```

```python
dict_5char['TGGTG']
```

```python
#ถ้าให้หา max min
maxVal = -100 #กำหนด max = ค่าน้อย / min = ค่ามาก
maxKey = None
for k in dict_5char.keys():  #วนลูป dict จำ
    if dict_5char[k] > maxVal:   #ถ้าค่า k > เป็นตัวนี้เลย แต่ถ้าน้อยจะไม่เก็บ เพราะเก็บตัวมากสุด 
        maxVal = dict_5char[k]
        maxKey = k
```

```python
maxKey
```

```python
print(f'max 5-long-sequence is {maxKey} with frequency equals {maxVal}')
```
