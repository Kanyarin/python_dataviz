# scatter plot
#### การกระจาย (ส่วนใหญ่ plot แบบ 2 มิติ)

```python
irisdata = pandas.read_csv('C:\\nutnoee\\python\\python_dataviz\\iris.data',header=None)   #การโหลดไฟล์นี้แบบไม่มีหัวตาราง
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()  #-1=ชื่อของคลาส
setosa_np[:5,:] ## .iloc คือจัดการด้วยตัวเลข / หน้า : ถ้าไม่ใส่อะไร แสดงว่าเอาตั้งแต่แถวแรก
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1]) #พารามิเตอร์ตัวที่ 1 และ 2
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=6)   #S=ขนาดของจุด
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'r')
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif 1 < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap ='YlOrBr') #สีเข้มค่ามาก สีอ่อนค่าน้อย
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = np.array([[1,0.5,0.8]])) 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr',marker='p')
```

# Visualize 2 data พ็อต 2 ข้อมูลในกราฟเดียว

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='BuPu',marker='p')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2], c = versicolor_np[:,3],cmap='BuPu',marker='*')
```

# Visualize 3 data เพื่อดูข้อมูล 3 อันว่าเหมือนหรือแตกต่างกันอย่างไร


```python
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = setosa_np[:,3],cmap='BuPu',marker="p")
plt.scatter(versicolor_np[:,0],versicolor_np [:,1],s=30*versicolor_np[:,2],c = versicolor_np[:,3],cmap='BuPu',marker="*" )
plt.scatter(virginica_np[:,0],virginica_np[:,1],s=30*virginica_np[:,2],c = virginica_np[:,3],cmap='BuPu',marker="x" )
```
